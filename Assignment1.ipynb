```python
#Histogram
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = pd.read_csv("used_cars_data.csv")

plt.figure(figsize=(10, 5))
plt.hist(data['Price'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Price (in lakhs)')
plt.ylabel('Number of Cars')
plt.title('Distribution of Car Prices')
plt.grid(False)
plt.show()
print(data.info)
```

```python
#Bar Chart
car_names = data['Name'].head(20)
car_prices = data['Price'].head(20)

plt.figure(figsize=(12, 6))
plt.barh(car_names, car_prices, color='skyblue')
plt.xlabel('Price (in lakhs)')
plt.ylabel("Cars")
plt.title('Car vs Price') 
plt.show()
```

```python
#Pie chart
Location_counts = data['Location'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(Location_counts, labels=Location_counts.index, autopct='%1.1f%%', startangle=90, shadow=True )
plt.title('Distribution of Cars in terms of Location')
plt.show()

fuel_counts = data['Fuel_Type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=90, shadow=True )
plt.title('Distribution of Fuel Type')
plt.show()
```

```python
#Scatter Plot
data['Engine'] = data['Engine'].str.replace(' CC', '', regex=False)
data['Engine'] = pd.to_numeric(data['Engine'], errors='coerce')

data_clean = data.dropna(subset=['Engine', 'Price'])

top10 = data_clean.head(50)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Engine', y='Price', data=top10, color='teal')
plt.xlabel('Engine (CC)')
plt.ylabel('Price (in lakhs)')
plt.title('Scatter Plot: Top 50 Cars - Engine vs Price')
plt.grid(True)
plt.show()



```

```python
#Bar Plot
data_clean = data.dropna(subset=['Price'])
avg_price_by_fuel = data_clean.groupby('Fuel_Type')['Price'].mean()

plt.figure(figsize=(8, 6))
plt.bar(avg_price_by_fuel.index, avg_price_by_fuel.values, color='coral')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price (in lakhs)')
plt.title('Fuel Type vs Average Car Price')
plt.show()
```
